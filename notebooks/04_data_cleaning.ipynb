```python
txn = pd.read_csv("../data/08_investor_transactions.csv")

print(txn.shape)
print(txn.columns.tolist())
```

```python
print("Missing Values:")
print(txn.isnull().sum())

print("\nDuplicate Rows:")
print(txn.duplicated().sum())

print("\nData Types:")
print(txn.dtypes)
```

```python
# Convert transaction date to datetime
txn['transaction_date'] = pd.to_datetime(txn['transaction_date'])

# Standardize text columns
text_cols = ['transaction_type', 'state', 'city', 'city_tier',
             'age_group', 'gender', 'payment_mode', 'kyc_status']

for col in text_cols:
    txn[col] = txn[col].astype(str).str.strip()

# Check for invalid transaction amounts
invalid_amounts = txn[txn['amount_inr'] <= 0]

print("Invalid transaction amounts:", len(invalid_amounts))

txn.head()
```

```python
txn.to_csv("../data/cleaned_investor_transactions.csv", index=False)

print("cleaned_investor_transactions.csv saved successfully")
```
