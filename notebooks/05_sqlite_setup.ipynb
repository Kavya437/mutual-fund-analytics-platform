```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("../data/mutual_fund.db")

print("Database created successfully")
```

```python
nav = pd.read_csv("../data/cleaned_nav_history.csv")
txn = pd.read_csv("../data/cleaned_investor_transactions.csv")

print(nav.shape)
print(txn.shape)
```

```python
nav.to_sql("nav_history", conn, if_exists="replace", index=False)
txn.to_sql("investor_transactions", conn, if_exists="replace", index=False)

print("Tables loaded successfully")
```

```python
query = """
SELECT name
FROM sqlite_master
WHERE type='table';
"""

pd.read_sql(query, conn)
```

```python
pd.read_sql("""
SELECT COUNT(*) as total_transactions
FROM investor_transactions
""", conn)
```

```python
pd.read_sql("""
SELECT COUNT(*) as total_nav_records
FROM nav_history
""", conn)
```

```python
import pandas as pd

fund = pd.read_csv("../data/01_fund_master.csv")
scheme = pd.read_csv("../data/07_scheme_performance.csv")

print(fund.shape)
print(scheme.shape)

print(fund.columns.tolist())
print(scheme.columns.tolist())
```

```python
fund.to_sql("fund_master", conn, if_exists="replace", index=False)

scheme.to_sql("scheme_performance", conn, if_exists="replace", index=False)

print("Additional tables loaded successfully")
```

```python
pd.read_sql("""
SELECT name
FROM sqlite_master
WHERE type='table'
""", conn)
```

```python
pd.read_sql("""
SELECT
    category,
    COUNT(*) as schemes
FROM fund_master
GROUP BY category
ORDER BY schemes DESC
""", conn)
```

```python
pd.read_sql("""
SELECT
    fund_house,
    AVG(return_3yr_pct) as avg_return_3yr
FROM scheme_performance
GROUP BY fund_house
ORDER BY avg_return_3yr DESC
LIMIT 10
""", conn)
```
