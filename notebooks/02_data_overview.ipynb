```python
import pandas as pd
import os

summary = []

for file in os.listdir("../data"):
    if file.endswith(".csv"):
        df = pd.read_csv(f"../data/{file}")

        summary.append({
            "Dataset": file,
            "Rows": df.shape[0],
            "Columns": df.shape[1],
            "Missing Values": df.isnull().sum().sum()
        })

summary_df = pd.DataFrame(summary)
summary_df
```
